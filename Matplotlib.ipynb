# Axes label, legend, grid:


```python
number = [1,3,4,6,8,9]
max_num = [80,70,85,90,76,69]
min_num = [28,51,34,45,31,40]
avg_num = [50,56,62,59,60,61]
plt.plot(number,min_num, label = "min",color = "red")
plt.plot(number,max_num, label = "max")
plt.plot(number,avg_num, label = "avg")
plt.legend(loc = "best",shadow = True)
plt.grid()
```

# Bar Chart:

```python
age =np.array([1,2,4,6,7])
population = [12,23,29,35,30]
edu = [40,30,45,25,60]
plt.bar(age-.3,population,label="pop",width =.4,color = "green")
plt.bar(age+.3,edu,label = "edu",width =.3, color="magenta")
plt.legend()
plt.grid()
```

# Pie Chart:

```python
x = [20,38,10,70,30,56]
l = ['jug','glass','mug','plate','lid','bottle']
plt.pie(x,labels =l, shadow =True, autopct = '%.3f%%',startangle= 30,explode=[0,.2,0,0,0,.08],radius = 1.5)
plt.savefig('piechart.pdf',bbox_inches= "tight", pad_inches =2, transparent=True)
```
